importing pandas library to read csv file from directory
then pass the data from csv into variable

```python
import pandas as pd

df = pd.read_csv('housepricedata.csv')
```

print all value in variable

```python
df
```

pass the value to new variable so when we make changes to the value, it will not affect the original variable and can be reused

```python
dataset = df.values
```

print the variable to see value available

```python
dataset
```

Pass all value from column 1 until 10 to new variable as a INPUT data that will be used to train the machine learning model.

```python
X = dataset[:,0:10]
```

```python
X
```

Pass all value in column 11 into new variable as OUTPUT data

```python
Y = dataset[:,10]
```

```python
Y
```

Import sklearn library to be used to preprocess the input data to Transforms input by scaling each feature to a given range of 0 and 1
This estimator scales and translates each feature individually such that it is in the given range on the training set.

```python
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
```

```python
X_scale
```

After finish preprocess the input data into new form, then we split the input into two part with ratio of 7:3 where it will be used for training the machine learning model and evaluate the model performance

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val_and_test, Y_train, Y_val_and_test = train_test_split(X_scale, Y, test_size=0.3)
```

```python
X_val, X_test, Y_val, Y_test = train_test_split(X_val_and_test, Y_val_and_test, test_size=0.5)
```

```python
print(X_train.shape, X_val.shape, X_test.shape, Y_train.shape, Y_val.shape, Y_test.shape)
```

Now we import keras library to build the machine learning model 

```python
from keras.models import Sequential
from keras.layers import Dense
```

for this, we build simple model called Deep neural network or known as DNN with 3 layer. this model used the relu activation function and accept input with shape of (10,1)

```python
model = Sequential([
    Dense(32, activation='relu', input_shape=(10,)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid'),
])
```

```python
model.summary()
```

We then compile the model with optimizer of sgd (Stochastic gradient descent), and set the loss to binary crossentropy. Binary crossentropy is a loss function used on problems involving yes/no (binary) decisions. For instance, in multi-label problems, where an example can belong to multiple classes at the same time, the model tries to decide for each class whether the example belongs to that class or not. Then we set the metric of the model into accuracy so the final result of training and evaluation will be inform of the accuracy

```python
model.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

Here we start training the model with batch size of 32 and epoch of 100 and pass the result into variable name 'hist'

```python
hist = model.fit(X_train, Y_train,
          batch_size=32, epochs=100,
          validation_data=(X_val, Y_val))
```

evaluate the model accuracy

```python
model.evaluate(X_test, Y_test)[1]
```

import matplotlib library to plot accuracy and loss graph

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper right')
plt.show()
```

```python
plt.plot(hist.history['acc'])
plt.plot(hist.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='lower right')
plt.show()
```
